# G1 Generador de Frases Aleatorias

Este generador crea frases aleatorias combinando elementos de tres listas. 
Selecciona al azar un elemento de cada lista para formar frases.

## Importar Librer√≠as


- `random`: permite seleccionar elementos al azar de listas, aqui para construir frases diferentes cada vez.
- `os`: maneja rutas y directorios del sistema, aqui para crear la carpeta de salida y definir la ruta del archivo.

```python
import random
import os
```

## Establecer listas

Definir listas de sujetos, verbos y complementos

```python
sujetos = [
    "El gato",
    "La estudiante",
    "Un programador",
    "La luna",
    "El robot",
]
verbos = [
    "observa",
    "construye",
    "descubre",
    "imagina",
    "escribe",
]
complementos = [
    "un misterio en el jardin.",
    "una idea brillante.",
    "la solucion del problema.",
    "un mensaje secreto.",
    "una historia curiosa.",
]
```

Definimos cuantas frases se van a generar y creamos la lista donde se almacenaran.

```python
num_frases = 5
frases = []
```

## Generar frases

Genera las frases combinando aleatoriamente un sujeto + verbo + complemento.

```python
for i in range(num_frases):
    sujeto = random.choice(sujetos)
    verbo = random.choice(verbos)
    complemento = random.choice(complementos)
    
    frase = f"{sujeto} {verbo} {complemento}"
    frases.append(frase)
    
    print(f"Frase {i + 1}: {frase}")
```

## Guardado frases generadas

Definir la ruta del archivo de salida.

```python
ruta_salida = os.path.join("output", "G1-Frases_Generadas.txt")
os.makedirs(os.path.dirname(ruta_salida), exist_ok=True)
```

Guardamos las frases en un archivo de texto `.txt`.

Se escribe cada frase en una linea del archivo.

```python
with open(ruta_salida, "w", encoding="utf-8") as archivo:
    for frase in frases:
        archivo.write(frase + "\n")

print(f"Frases guardadas en: {ruta_salida}")
```
