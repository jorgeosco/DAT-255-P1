# G2 - Generador de Imágenes


Genera un lienzo con figuras geométricas de vértices aleatorios usando NumPy y 
Matplotlib, asigna color de forma aleatoria a cada figura, los dibuja en un área 
configurable.

## Importar Librerías

- `NumPy`: biblioteca para arrays y operaciones numéricas eficientes (genera coordenadas aleatorias).
- `matplotlib.pyplot`: librería espacializada en la creación de gráficos y visualizaciones.
- `matplotlib.patches`: modulo que contiene clases para crear y maniputar formas geométricas 2D, como el `Polygon` que se usa para dibujar los poligonos.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

## Parametros generales

Se define el ancho y alto del lienzo.
Tambien se establece el numero de figuras, el numero de vertices por figura
y una paleta simple de colores.

```python
ancho = 10
alto = 10

num_figuras = 7
num_vertices = 3
colores = ['r', 'g', 'b', 'y', 'c', 'm']
```

## Preparación del lienzo
Prepara el lienzo y el area de dibujo para agregar los poligonos.

Crea dos objetos: 
- `fig` es la figura (el contenedor del grafico).
- `ax` es el eje donde se dibuja. En otras palabras, 

```python
fig, ax = plt.subplots()
```

## Generacion de poligonos aleatorios

Para cada figura:
1. Se generan coordenadas `x` y `y` aleatorias dentro del lienzo.
2. Se combinan en una lista de vertices.
3. Se crea un poligono con color y transparencia aleatoria.
4. Se agrega el poligono a los ejes.

Nota: `alpha` controla la transparencia en el rango [0, 1].

```python
for i in range(num_figuras):
    # Cada figura es un polígono con num_vertices vértices en posición aleatoria
    x = np.random.rand(num_vertices) * ancho
    y = np.random.rand(num_vertices) * alto
    vertices = list(zip(x, y)) # lista de coordenadas

    # Crear el polígono con color y transparencia aleatorios
    figura = Polygon(
        vertices,
        closed=True,
        alpha=np.random.rand(),
        facecolor=np.random.choice(colores),
    )
    ax.add_patch(figura)
```

## Ajuste de la visualizacion

Se fija el rango de los ejes, la proporcion del aspecto y se ocultan las
marcas de los ejes para una salida limpia.

```python
ax.set_xlim(0, ancho)
ax.set_ylim(0, alto)
ax.set_aspect('equal', adjustable='box')
ax.set_xticks([])
ax.set_yticks([])
```

## Guardado de la imagen

Se guarda la figura en un archivo de imagen `.png`.

```python
ruta_salida = "output/G2-figuras_aleatorias.png"
plt.savefig(ruta_salida)
print(f"Imagen guardada en: {ruta_salida}")
```

## Ejecución

```python
# Crear la figura, generar los polígonos y configurar el lienzo
fig, ax = plt.subplots()

# Generar y agregar cada figura al lienzo
for i in range(num_figuras):
    # Cada figura es un polígono con num_vertices vértices en posición aleatoria
    x = np.random.rand(num_vertices) * ancho
    y = np.random.rand(num_vertices) * alto
    vertices = list(zip(x, y))

    # Crear el polígono con color y transparencia aleatorios
    figura = Polygon(
        vertices,
        closed=True,
        alpha=np.random.rand(),
        facecolor=np.random.choice(colores),
    )
    ax.add_patch(figura)

# Configurar los límites y apariencia del lienzo
ax.set_xlim(0, ancho)
ax.set_ylim(0, alto)
ax.set_aspect('equal', adjustable='box')
ax.set_xticks([])
ax.set_yticks([])

# Guardar la imagen generada a disco (modificar la ruta si se desea)
ruta_salida = "output/G2-figuras_aleatorias.png"
plt.savefig(ruta_salida)
print(f"Imagen guardada en: {ruta_salida}")

plt.show()
```
